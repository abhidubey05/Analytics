```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load datset
df=pd.read_csv('Housing.csv')
df
```

```python
print(df.head())
```

```python
# box plot
sns.boxplot(data=df, x='bedrooms', y='price', palette='Set2')
plt.title('House Price Distribution by Bedrooms')
plt.show()
```

```python
# count plot
sns.countplot(data=df, x='furnishingstatus', palette='pastel')
plt.title('Count of Houses by Furnishing Status')
plt.show()
```

```python

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df)
plt.suptitle("Pair Plot of All Numeric Features", y=1.02)
plt.show()
```

```python
sns.pairplot(df, hue='furnishingstatus')
plt.suptitle("Pair Plot by Furnishing Status", y=1.02)
plt.show()
```

```python
sns.lmplot(data=df, x='area', y='price', height=6, aspect=1.5)
plt.title("Linear Regression: Area vs Price")
plt.show()
```

```python
sns.violinplot(data=df, x='bedrooms', y='price', palette='muted')
plt.title("Violin Plot: Price by Bedrooms")
plt.show()
```

```python
sns.stripplot(data=df, x='bedrooms', y='price', jitter=True, palette='Set1')
plt.title("Price Distribution Across Bedrooms (Strip Plot)")
plt.show()
```

```python
sns.swarmplot(data=df, x='furnishingstatus', y='price', palette='Set2')
plt.title("Price Distribution by Furnishing Status (Swarm Plot)")
plt.show()
```
